```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/content/netflix_titles.csv (2).xlsx")
print("Initial Dataset Shape:", df.shape)
print(df.info())
print("Missing Values per Column:\n", df.isnull().sum())
```

```python
# Remove Duplicates
df.drop_duplicates(inplace=True)

# Handle Missing Values
df['director'].fillna('Not Available', inplace=True)
df['cast'].fillna('Not Available', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna(df['rating'].mode()[0], inplace=True)
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Standardize Text Values
df['country'] = df['country'].str.strip().str.title()
df['type'] = df['type'].str.strip().str.title()
df['rating'] = df['rating'].str.strip().str.upper()

# Rename Columns
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Convert Date Format to dd-mm-yyyy
df['date_added'] = df['date_added'].dt.strftime('%d-%m-%Y')

# Check and Fix Data Types
df['date_added'] = pd.to_datetime(df['date_added'], format='%d-%m-%Y', errors='coerce')
df['duration_mins'] = df['duration'].str.extract(r'(\\d+)').astype(float)

print("\nData Types After Cleaning:\n", df.dtypes)
```

```python
df.to_csv("netflix_cleaned.csv", index=False)
files.download("netflix_cleaned.csv")
```
